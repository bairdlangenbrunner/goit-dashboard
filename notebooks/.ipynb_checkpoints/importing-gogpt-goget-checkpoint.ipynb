## calculations for gogpt data

```python
gas_plants_af1 = gas_plants_ssa.loc[gas_plants_ssa.Region=='Africa']
gas_plants_af2 = gas_plants_mena.loc[gas_plants_mena.Region=='Africa']
gas_plants_africa_df = pandas.concat([gas_plants_af1, gas_plants_af2])

country_list = sorted(gas_plants_africa_df.Country.unique().tolist())

gas_plants_capacity_sum = pandas.DataFrame(0, index=country_list, columns=['Announced',
                                                                           'Pre-construction',
                                                                           'Construction'])

# make any missing capacities NaN
gas_plants_africa_df['TOTAL Capacity elec. (MW)'] = gas_plants_africa_df['TOTAL Capacity elec. (MW)'].astype(str)
gas_plants_africa_df['TOTAL Capacity elec. (MW)'] = gas_plants_africa_df['TOTAL Capacity elec. (MW)'].str.replace(',','')
gas_plants_africa_df['TOTAL Capacity elec. (MW)'].replace(['','not found'],numpy.nan,inplace=True)
gas_plants_africa_df['TOTAL Capacity elec. (MW)'] = gas_plants_africa_df['TOTAL Capacity elec. (MW)'].astype(float)

status_rename = {'announced':'Announced',
                 'cancelled':'Cancelled',
                 'construction':'Construction',
                 'mothballed':'Mothballed',
                 'operating':'Operating',
                 'pre-constructoin':'Pre-construction',
                 'retired':'Retired',
                 'shelved':'Shelved'}

#gas_plants_africa_df.groupby(['Country','Status'])['TOTAL Capacity elec. (MW)'].sum().unstack().replace(numpy.nan,0)

gas_plants_capacity_sum['Announced'] += gas_plants_africa_df.loc[gas_plants_africa_df.Status=='announced'].groupby(
        'Country')['TOTAL Capacity elec. (MW)'].sum()
gas_plants_capacity_sum['Pre-construction'] += gas_plants_africa_df.loc[gas_plants_africa_df.Status=='pre-construction'].groupby(
        'Country')['TOTAL Capacity elec. (MW)'].sum()
gas_plants_capacity_sum['Construction'] += gas_plants_africa_df.loc[gas_plants_africa_df.Status=='construction'].groupby(
        'Country')['TOTAL Capacity elec. (MW)'].sum()

gas_plants_capacity_sum.replace(numpy.nan,0,inplace=True)
```

## import goget data

looks like there's two sheets, a main data sheet and a production/reserves sheet

the main sheet has all projects, the production sheet only has those projects that have production data

* use main sheet to move through unit ids
  * for each unit id in main sheet, collect anything that's gas in the production/reserves sheet
  * do a sum of that gas m^3, store it in main sheet

```python
spreadsheet = gc.open_by_key('1sJj5EvU1F9F2eTNvqyAFgEK_953kSh0RdbfOGZd78Co') # gogpt from may 2023

goget_repro_df = spreadsheet.worksheet('title', 'Reserves and Production').get_as_df()
goget_main_df = spreadsheet.worksheet('title', 'Main data').get_as_df()
goget_regions_df = spreadsheet.worksheet('title', 'Country Names and Region Definitions').get_as_df()

# remove any unused unit ids by filtering out the rows without a unit name
goget_main_df = goget_main_df.loc[goget_main_df['Unit Name']!='']
goget_repro_df = goget_repro_df.loc[goget_repro_df['Unit Name']!='']

# set unit id to the index for goget main
goget_main_df.set_index('Unit ID', inplace=True)
```

```python
goget_regional_countries = goget_regions_df.loc[goget_regions_df['GEM region'].isin(['Africa'])]
goget_repro_region_df = goget_repro_df.loc[goget_repro_df.Country.isin(goget_regional_countries['GEM Standard Country Name'].tolist())]
goget_main_region_df = goget_main_df.loc[goget_main_df.Country.isin(goget_regional_countries['GEM Standard Country Name'].tolist())]
unit_id_list = goget_main_region_df.index.unique().tolist()

reserves_sum = pandas.DataFrame(0, index=country_list, columns=['Discovered',
                                                                'In development'])

goget_repro_region_gas_df = goget_repro_region_df.loc[goget_repro_region_df['Fuel description'].str.contains('gas')]

#add a status column to repro corresopnding to main
goget_main_region_df.loc[:,'Gas volume'] = numpy.nan
for idx,row in goget_repro_region_gas_df.iterrows():
    if row['Unit ID'] in unit_id_list:
        pass
        goget_repro_subset = goget_repro_region_gas_df.loc[goget_repro_region_gas_df['Unit ID']==row['Unit ID']]
        goget_main_region_df.loc[row['Unit ID'],'Gas volume'] = goget_repro_subset['Quantity (converted)'].sum(min_count=0)
    else:
        print('missing')
```

```python
reserves_sum['Discovered'] = goget_main_region_df.groupby(['Country','Status'])['Gas volume'].sum().unstack()[['discovered']]#[['in development']]
reserves_sum['In development'] = goget_main_region_df.groupby(['Country','Status'])['Gas volume'].sum().unstack()[['in development']]
reserves_sum.replace([numpy.nan],0,inplace=True)
```

```python
reserves_count = pandas.DataFrame(0, index=country_list, columns=['Discovered',
                                                                'In development'])

reserves_count['Discovered'] = goget_main_region_df.groupby(['Country','Status'])['Gas volume'].count().unstack()[['discovered']]#[['in development']]
reserves_count['In development'] = goget_main_region_df.groupby(['Country','Status'])['Gas volume'].count().unstack()[['in development']]
reserves_count.replace([numpy.nan],0,inplace=True)
```
